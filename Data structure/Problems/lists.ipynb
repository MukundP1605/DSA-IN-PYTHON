### List problems

```python
### You are given a list of integers.

### Your task is to calculate and print the sum of all elements present in the list.

n = [1,2,3,4,5]
s = 0
for i in n:
    s += i
print(s)
```

```python
## You are given a list of integers.

## Your task is to find and print the largest element present in the list.

largest = n[0]

for i in n :
    if i > largest:
        largest = i 
print(largest)
```

```python
## remove duplicate from the list 

n = [1,2,2,3,4,3,5]
s = []
seen = set()
for i in n:
    if i not in seen :
        s.append(i)
        seen.add(i)
print(s)
```

```python
## check whether all elements in the list are unique.

n = [1,2,2,3,4,3,5]
unique = True
seen = set()

for i in n:
    if i in seen:
        unique = False
        break
    seen.add(i)

print(unique)
    
    
```
