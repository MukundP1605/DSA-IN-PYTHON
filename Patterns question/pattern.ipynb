Problem 1: Print Solid Square

Problem Statement:
You are given an integer N. Print a solid square of side length N using the * character.

Input Format:

A single integer N

Output Format:

Print N lines

Each line contains exactly N * characters separated by a single spac

```python
n = int(input())
i = 0 
while i < n:
    print("* "*n)
    i += 1

```

```python
""" You are given an integer N representing the number of rows in the upper half of a diamond.

Your task is to print a diamond-shaped pattern using the * (asterisk) character.

The diamond must be vertically and horizontally symmetric

The widest part of the diamond appears in the middle

The total number of rows printed should be 2 × N − 1"""

n = int(input())

for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)
    
for i in range(n -1 , 0 , -1):
    print(" " * (n - i) + "* " * i)
```
