### binary search


```python
## binary search 
arr = [1,2,3,4,5,6,7,8,9]
target = 7
n = len(arr)
start = 0
end = n - 1

while start <= end :
    mid = start + (end - start)//2
    if arr[mid] == target :
        print(f"element found at index {mid}"  )
        break
    elif arr[mid] > target :
        end = mid - 1
    else :
        start = mid +1
else :
    print("element not found")
```
