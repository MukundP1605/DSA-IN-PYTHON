```python
###  Find minimum and maximum element in an array

arr = [34,43,73,235,8,33,398,2,7846,3,64234,63,553,0]
minimum = arr[0]


maximum = arr[0]

for i in arr:
    if i > maximum :
        
        maximum = i
        
    if i < minimum :
        minimum = i   

print(minimum)    

print(maximum)
```

```python
### second largest element in the array
arr = [34,43,73,235,8,33,398,2,7846,3,64234,63,553,0]
sl = -1
a = -1

for i in arr :
    if i > a:
        sl = a
        a = i
    elif i < a and i > sl :
        sl = i
print(sl)

```

```python
arr = [1,2,2,3,1,3,4]
frq = {}

for i in arr:
    if i in frq:
        frq[i]+=1
    else :
        frq[i]=1
print(frq)
```
