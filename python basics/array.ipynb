```python
###  Find minimum and maximum element in an array

arr = [34,43,73,235,8,33,398,2,7846,3,64234,63,553,0]
minimum = arr[0]
maximum = arr[0]

for i in arr:
    if i > maximum :
        maximum = i
    if i < minimum :
        minimum = i
    

print(minimum)    
print(maximum)
```
