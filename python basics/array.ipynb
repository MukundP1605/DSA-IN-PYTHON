```python
###  Find minimum and maximum element in an array

arr = [34,43,73,235,8,33,398,2,7846,3,64234,63,553,0]
minimum = arr[0]


maximum = arr[0]

for i in arr:
    if i > maximum :
        
        maximum = i
        
    if i < minimum :
        minimum = i   

print(minimum)    

print(maximum)
```

## Two Pointers

```python
## 1 Write a function that reverses a array. The input array is given as an array .

## You must do this by modifying the input array in-place with O(1) extra memory.

arr = [1,2,3,4,5,6]
n = len(arr)

left = 0
right = n - 1
if n == 0:
    print("array is empty")
else :
    while right > left:
        arr[right],arr[left] = arr[left],arr[right]
        right-=1
        left +=1
    print(arr)

```

```python
## Two Sum II â€“ Input Array Is Sorted

arr = [1,2,7,11,15]
n = len(arr)
target = 22

for i in range(0,n-1):
    for j in range(i+1,n):
        if (arr[i] + arr[j]) == target:
            print(i+1,j+1)
            break
```

```python
## two pointer approch
def ds(arr):
    arr = [1,2,7,23,24]
    target = 25

    left , right = 0 , len(arr)-1

    while right>left:
        if arr[left] + arr[right] == target:
            return (left+1,right+1)
            break
        elif arr[left] + arr[right] > target:
            right-=1
        else :
            left +=1
    return -1
result = ds(arr)
print(result)
```

```python
### Problem: Check if an Array is a Palindrome

arr = [1,2,3,2,1]
left , right = 0 , len(arr) - 1
pl = False
while right > left :
    if arr[left] == arr[right]:
        left +=1 
        right -=1
        pl = True
    else :
        pl = False
        
print(pl)
            
        
```
